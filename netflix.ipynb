```python
df['director'].fillna('Unknown', inplace= True)
df['cast'].fillna('Unknown', inplace= True)
df['country'].fillna('Unknown', inplace= True)
df['date_added'].fillna('Unknown', inplace= True)
df['rating'].fillna('Unknown', inplace= True)
df['duration'].fillna('Unknown', inplace= True)
df.head(50)
```

```python
#df['rating'].value_counts()
df['country'].value_counts().head(10)
```

```python
#Identifying similar content by matching text-based features
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# Combine text features
df['text_features'] = df['description'] + ' ' + df['listed_in']

# Convert text to vectors
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['text_features'])

# Compute similarity between all movies/shows
similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Example: find most similar shows to the first one
similar_movies = similarity_matrix[0]                  # similarity scores for first movie
top_indices = similar_movies.argsort()[::-1][1:6]     # sort descending, skip itself
df.iloc[top_indices][['title', 'type', 'listed_in']]  # show top 5 similar
```

```python
#Does Netflix has more focus on TV Shows than movies in recent years
#count movies and TV shows per year
content_per_year= df.groupby(['release_year','type']).size().unstack(fill_value=0)
print(content_per_year.head(43))
```

```python
#Visualizing
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(content_per_year.index, content_per_year['Movie'], label='Movies')
plt.plot(content_per_year.index, content_per_year['TV Show'], label='TV Shows')
plt.xlabel('Year of Release')
plt.ylabel('Number of Titles')
plt.title('Netflix: Movies vs TV Shows Over Years')
plt.legend()
plt.show()
```
